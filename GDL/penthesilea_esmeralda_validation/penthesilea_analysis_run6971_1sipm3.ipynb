```python
import glob
import tables as tb
import numpy  as np
import pandas as pd

import matplotlib.pyplot as plt

from histogram_functions import *

to_df = pd.DataFrame.from_records
```

```python
run = 6971
trigger="trigger1"

icversion_datadir  = f"/home/jrenner/analysis/{run}/hdf5/1sipm_3/hdst_v099/{trigger}/" 
testprod_datadir   = f"/home/jrenner/analysis/{run}/hdf5/1sipm_3/hdst_IC-testprod/{trigger}/"  
testprod_p_datadir = f"/home/jrenner/analysis/{run}/hdf5/1sipm_3/hdst_IC-testprod-p/{trigger}/"
```

```python
icversion_files = glob.glob( icversion_datadir + "*")
icversion_files.sort()

testprod_files = glob.glob( testprod_datadir + "*")
testprod_files.sort()

testprod_p_files = glob.glob( testprod_p_datadir + "*")
testprod_p_files.sort()
```

```python
def create_ntuple(files, nfiles):
    df = pd.DataFrame([])
    i=1
    for file in files:
        hdf5 = tb.open_file(file) 
        df = pd.concat( [df, to_df( hdf5.root.DST.Events.read())])
        
        hdf5.close()
        if i>=nfiles: break
        i+=1            
    return df
```

```python
nfiles=300
icversion_dst  = create_ntuple(icversion_files , nfiles)
testprod_dst   = create_ntuple(testprod_files  , nfiles)
testprod_p_dst = create_ntuple(testprod_p_files, nfiles) 
```

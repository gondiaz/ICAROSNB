## Charge per Slice per SIPM

```python
sipm_charge = testprod_p_file.root.SIPM.sipm_charge.read()
```

```python
sipms = np.unique(sipm_charge[:, 0])

cut=6e3
for sipm in sipms:
    sel = sipm_charge[:, 0]==sipm
    data = sipm_charge[sel][:, 1]
    
    if len(data)>=cut:
        histo1D(data,
                bins=100,
                xlabel="charge",
                xlims=[0, 80],
                Label0=f"{sipm}",
                scale="log")
```

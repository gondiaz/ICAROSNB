```python
print("1S1 1S2 RATES")
print()
print("IC v0.9.9"    , len(nS_icversion[(nS_icversion["nS1s"]==1)&(nS_icversion["nS2s"]==1)])/len(nS_icversion)*100)
print("IC testprod"  , len(nS_testprod[(nS_testprod["nS1s"]==1)&(nS_testprod["nS2s"]==1)])/len(nS_testprod)*100)
print("IC testprod_p", len(nS_testprod_p[(nS_testprod_p["nS1s"]==1)&(nS_testprod_p["nS2s"]==1)])/len(nS_testprod_p)*100)
```

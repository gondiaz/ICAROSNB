```python
import csth.utils.fits as efit

def plot_extll_fit(es, e0, e1, mushat, nbins = 100, alpha = 0.5, tag = ''):
    pg = efit.Pol1DGaus(a = e0, b = e1, shapes = 'mu, sigma, mslope, f')
    mu, sigma, mslope, nug, nup = mushat[0], mushat[1], mushat[2], mushat[3], mushat[4]
    ntot = nug + nup
    f = float(nug)/float(nug+nup)
    pf = efit.Pol1DGaus(a = e0, b = e1, shapes = 'mu, sigma, mslope, f')
    xs = np.linspace(e0, e1, nbins)
    dx = xs[1]-xs[0]
    ys = ntot * dx * pg.pdf(xs, mu = mu, sigma = sigma, mslope = mslope, f = f)
    plt.hist(es, nbins, range = (e0, e1), alpha = alpha)
    fwhm = 235.* sigma/mu
    label = tag + ' n = ' + str(int(nug)) + ', $\mu$ = ', "{0:.0f}".format(mu) + ', FWHM (%) = ' + "{0:.2f}".format(fwhm)
    plt.plot(xs, ys, ls= '-', lw = 2, label = label, alpha = alpha)
    plt.xlim(e0, e1)
    plt.legend()
    return (xs, ys)
    
```

```python
def canvas_esigma(esum, erange, nbins= 100, tag='', alpha = 0.5):
    e0, e1 = erange
    dssel = in_range(esum.e, e0, e1)
    es = esum.e[dssel]
    mushat = efit.extll_fit_pol1dgaus(es, e0, e1)
    print(es.min(), es.max())
    #plt.hist(es, nbins, range = (360e3, 440e3))
    #plt.xlim(e0, e1)
    dat = plot_extll_fit(es, e0, e1, mushat, tag = tag, alpha = alpha)
    return mushat, dat
```

```python
pg = efit.Pol1DGaus(a = e0, b = e1, shapes = 'mu, sigma, mslope, f')
```

```python
def ll_fun(xs):
    def llfit(mus):
        mu, sigma, mslope, f = mus[0], mus[1], mus[2], mus[3]
        return -2.*np.sum(pg.logpdf(xs, mu = mu, sigma = sigma, mslope = mslope, f = f))
    return llfit
```

```python
es = esum.e[dssel]
mu, sigma, mslope, nug, nup = 410e3, 5e3, 0., 1000, 7000
mu, sigma, mslope, f = 410e3, 10e3, 0., 0.2
#res = efit.extll_fit_pol1gaus(es, e0, e1, mu, sigma, mslope, nug, nup)
```

```python
from scipy.optimize import minimize

llfit  = ll_fun(es) 
mus0   = np.array([mu, sigma, mslope, f])   # initial guess
res    = minimize(llfit, mus0, method='Nelder-Mead' )
mushat = res.x
```

```python
def canvas_eq_vs_var(varname, erange, alpha = 0.01):
    xsel = (in_range(esum.e, *erange)) & (in_range(esum.eq, 6, 14))
    c = hst.Canvas(3, 2)
    c(1)
    var = esum[varname].values[xsel]
    plt.scatter(var, esum.e [xsel], alpha = alpha)
    c(2)
    plt.hist2d (var, esum.e [xsel], (50, 50))
    xs, ys, eys = fitf.profileX(var, esum.e[xsel], 50, yrange = erange)
    hst.errorbar(xs, ys, yerr=eys, fmt='*', canvas=c(2), c='black')    
    c(3)
    plt.scatter(var, esum.q [xsel], alpha = alpha)
    c(4)
    plt.hist2d (var, esum.q [xsel], (50, 50))
    c(5)
    plt.scatter(var, esum.eq[xsel], alpha = alpha)
    c(6)
    plt.hist2d (var, esum.eq[xsel], (50, 50))
    return
    
```
